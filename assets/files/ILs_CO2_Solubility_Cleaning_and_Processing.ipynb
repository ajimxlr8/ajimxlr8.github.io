```python
import pandas as pd

# Load the dataset
# Skip the first few rows based on the preview
df = pd.read_excel('/content/drive/MyDrive/Matlab Project/Co2 solubility.xlsx', skiprows=3)

# Display basic info and the first few rows
print("--- Data Info ---")
print(df.info())
print("\n--- First 5 Rows ---")
print(df.head())
print("\n--- Value Counts for first column (to check header alignment) ---")
print(df.iloc[:, 0].value_counts().head())
```

```python
import numpy as np

# 1. Remove rows where IL-CO2 Systems is NaN
df_cleaned = df.dropna(subset=['IL-CO2 Systems']).copy()

# 2. Replace '-' with NaN across the whole dataframe
df_cleaned.replace('-', np.nan, inplace=True)

# 3. Convert relevant columns to numeric
cols_to_fix = [
    'Solubility(cal.)', 'RD1', 'Solubility(COSMO-RS(combi2005))',
    'RD1.1', 'Solubility(COSMO-RS(combi1998))', 'RD2'
]
for col in cols_to_fix:
    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')

# 4. Check for and drop duplicates
duplicates_count = df_cleaned.duplicated().sum()
df_cleaned.drop_duplicates(inplace=True)

# 5. Drop purely descriptive columns
df_cleaned.drop(columns=['Full References', 'No. of refs.'], inplace=True)

# Inspect missing values after conversion
print("--- Missing Values Count ---")
print(df_cleaned.isnull().sum())

# Inspect column names to ensure everything is clean
print("\n--- Column Names ---")
print(df_cleaned.columns.tolist())
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

# 1. Comprehensive Mapping Dictionary
cation_smiles = {
    '[(ETO)2IM]': 'CCOn1cc[n+](OCC)c1', '[b2-Nic]': 'CCCCOC(=O)c1ccc[n+](c1)CCCC', '[BBIM]': 'CCCCn1cc[n+](c1)CCCC',
    '[bhea]': 'OCC[NH2+]CCO', '[BMIM]': 'CCCCn1cc[n+](c1)C', '[BMPYR]': 'CCCC[N+]1(C)CCCC1', '[BPY]': 'CCCC[n+]1ccccc1',
    '[C2OMIM]': 'COCCn1cc[n+](c1)C', '[C3MPYR]': 'CCC[N+]1(C)CCCC1', '[C5MIM]': 'CCCCCn1cc[n+](c1)C', '[C5MPYR]': 'CCCCC[N+]1(C)CCCC1',
    '[C6F9MIM]': 'FC(F)(F)C(F)(F)C(F)(F)CCn1cc[n+](c1)C', '[C7MPYR]': 'CCCCCCC[N+]1(C)CCCC1', '[C8F13MIM]': 'FC(F)(F)C(F)(F)C(F)(F)C(F)(F)C(F)(F)C(F)(F)CCn1cc[n+](c1)C',
    '[DMFH]': 'CN(C)C=[NH2+]', '[DMIM]': 'Cn1cc[n+](c1)C', '[EMIM]': 'CCn1cc[n+](c1)C', '[Et3NBH2MIM]': 'CC[N+](CC)(CC)CCCP(=O)(O)O',
    '[ETT]': 'CC[P+](c1ccccc1)(c2ccccc2)c3ccccc3', '[he]': '[NH3+]CCO', '[hea]': '[NH3+]CCO', '[HEMIM]': 'OCCn1cc[n+](c1)C',
    '[hheme]': 'OCCn1c(C)c[n+](c1)C', '[HMIM]': 'CCCCCCn1cc[n+](c1)C', '[HMMIM]': 'CCCCCCn1c(C)c[n+](c1)C', '[HMPY]': 'CCCCCC[n+]1ccc(C)cc1',
    '[HMPYR]': 'CCCCCC[N+]1(C)CCCC1', '[HOC3MIM]': 'OCCCn1cc[n+](c1)C', '[m-2-HEA]': 'C[NH2+]CCO', '[MDEA]': 'C[NH+](CCO)CCO',
    '[MMIM]': 'Cn1cc[n+](c1)C', '[N1,1,1,2-OH]': 'C[N+](C)(C)CCO', '[N1,1,3,2-OH]': 'C[N+](C)(C)CC(C)O', '[N1,8,8,8]': 'C[N+](CCCCCCCC)(CCCCCCCC)CCCCCCCC',
    '[N2,1,1,3]': 'CC[N+](C)(C)CCC', '[N4,1,1,1]': 'CCCC[N+](C)(C)C', '[N1,4,4,4]': 'C[N+](CCCC)(CCCC)CCCC', '[N4,4,4,4]': 'CCCC[N+](CCCC)(CCCC)CCCC',
    '[NMIM]': 'c1c[nH]c[n+]1C', '[NMPYR]': 'C[N+]1CCCC1', '[OMIM]': 'CCCCCCCCn1cc[n+](c1)C', '[OMPYR]': 'CCCCCCCC[N+]1(C)CCCC1',
    '[P1,4,4,4]': 'C[P+](CCCC)(CCCC)CCCC', '[P4,4,4,4]': 'CCCC[P+](CCCC)(CCCC)CCCC', '[P6,6,6,14]': 'CCCCCC[P+](CCCCCC)(CCCCCC)CCCCCCCCCCCCCC',
    '[Py2OH]': 'OCC[n+]1ccccc1', '[thea]': 'OCC[NH+](CCO)CCO', '[TMG]': 'CN(C)C(=[NH2+])N(C)C'
}

anion_smiles = {
    '[Tf2N]': 'C(F)(F)(F)S(=O)(=O)[N-]S(=O)(=O)C(F)(F)F', '[MeSO3]': 'CS(=O)(=O)[O-]', '[Ac]': 'CC(=O)[O-]', '[L]': 'CC(O)C(=O)[O-]',
    '[BF4]': 'F[B-](F)(F)F', '[C3F7CO2]': 'FC(F)(F)C(F)(F)C(F)(F)C(=O)[O-]', '[C7F15CO2]': 'FC(F)(F)C(F)(F)C(F)(F)C(F)(F)C(F)(F)C(F)(F)C(F)(F)C(=O)[O-]',
    '[Cl]': '[Cl-]', '[DBPO4]': 'CCCCOP(=O)([O-])OCCCC', '[DCA]': 'N#C[N-]C#N', '[eFAP]': 'F[P-](F)(F)(C(F)(F)C(F)(F)F)(C(F)(F)C(F)(F)F)C(F)(F)C(F)(F)F',
    '[IAAc]': 'IC[C](=O)[O-]', '[ISB]': 'C1C(OC2C1OC(C(O2)O)O)[O-]', '[LEV]': 'CC(=O)CCC(=O)[O-]', '[MDEGSO4]': 'COCCOCCS(=O)(=O)[O-]',
    '[MeSO4]': 'COS(=O)(=O)[O-]', '[methide]': 'C(S(=O)(=O)C(F)(F)F)(S(=O)(=O)C(F)(F)F)[S-](=O)(=O)C(F)(F)F', '[NO3]': '[O-][N+](=O)[O-]',
    '[PF6]': 'F[P-](F)(F)(F)(F)F', '[Pro]': 'C1CCNC1C(=O)[O-]', '[SCN]': '[S-]C#N', '[SUC]': '[O-]C(=O)CCC(=O)[O-]', '[TFA]': 'FC(F)(F)C(=O)[O-]',
    '[TFES]': 'FC(F)C(F)(F)S(=O)(=O)[O-]', '[TfO]': 'FC(F)(F)S(=O)(=O)[O-]', '[TMA]': 'O=C([O-])c1cc(cc(c1)C(=O)[O-])C(=O)[O-]',
    '[IDA]': '[O-]C(=O)CNCC(=O)[O-]', '[bFAP]': 'F[P-](F)(F)(C(F)(F)C(F)(F)C(F)(F)C(F)(F)F)(C(F)(F)C(F)(F)C(F)(F)C(F)(F)F)C(F)(F)C(F)(F)C(F)(F)C(F)(F)F',
    '[DEPO4]': 'CCOP(=O)([O-])OCC', '[EtSO4]': 'CCOS(=O)(=O)[O-]', '[FOR]': '[H]C(=O)[O-]', '[ACE]': 'CC1=CC(=O)[N-]S(=O)(=O)O1',
    '[Br]': '[Br-]', '[pFAP]': 'F[P-](F)(F)(C(F)(F)C(F)(F)F)(C(F)(F)C(F)(F)F)C(F)(F)C(F)(F)F', '[SAC]': 'C1=CC=C2C(=C1)C(=O)[N-]S2(=O)=O',
    '[TCB]': '[B-](C#N)(C#N)(C#N)C#N', '[DMPO4]': 'COP(=O)([O-])OC', '[MP]': 'CP(=O)([O-])O', '[doc]': 'CCCCC(CC)COC(=O)CC(C(=O)OCC(CC)CCCC)S(=O)(=O)[O-]',
    '[TOS]': 'Cc1ccc(cc1)S(=O)(=O)[O-]', '[Ala]': 'CC(N)C(=O)[O-]', '[C12H25PhSO3]': 'CCCCCCCCCCCCc1ccc(cc1)S(=O)(=O)[O-]',
    '[Gly]': 'NCC(=O)[O-]', '[Ile]': 'CCC(C)C(N)C(=O)[O-]', '[Lys]': 'NCCCCC(N)C(=O)[O-]', '[phos]': '[O-]P(=O)([O-])[O-]',
    '[Sar]': 'CNCC(=O)[O-]', '[Tau]': 'NCCS(=O)(=O)[O-]'
}

def smiles_to_mol(smiles):
    """Safely convert SMILES to RDKit Mol"""
    if pd.isna(smiles):
        return None
    try:
        return Chem.MolFromSmiles(smiles)
    except:
        return None

def compute_descriptors(mol):
    """Return descriptor dict safely"""
    if mol is None:
        return {
            "MW": None,
            "LogP": None,
            "TPSA": None
        }
    return {
        "MW": Descriptors.MolWt(mol),
        "LogP": Descriptors.MolLogP(mol),
        "TPSA": Descriptors.TPSA(mol)
    }

def add_rdkit_features(df_cleaned, cation_smiles, anion_smiles):
    # --- Map SMILES ---
    df_cleaned['cat_smiles'] = df_cleaned['Cation'].map(cation_smiles)
    df_cleaned['ani_smiles'] = df_cleaned['Anion'].map(anion_smiles)

    # --- Convert to molecules ONCE ---
    df_cleaned['cat_mol'] = df_cleaned['cat_smiles'].apply(smiles_to_mol)
    df_cleaned['ani_mol'] = df_cleaned['ani_smiles'].apply(smiles_to_mol)

    # --- Compute descriptors ---
    cat_desc = df_cleaned['cat_mol'].apply(compute_descriptors).apply(pd.Series)
    ani_desc = df_cleaned['ani_mol'].apply(compute_descriptors).apply(pd.Series)

    # --- Rename columns ---
    cat_desc = cat_desc.add_prefix('cat_')
    ani_desc = ani_desc.add_prefix('ani_')

    # --- Merge back ---
    df_cleaned = pd.concat([df_cleaned, cat_desc, ani_desc], axis=1)

    # --- Optional: drop intermediate mol objects ---
    df_cleaned.drop(columns=['cat_mol', 'ani_mol'], inplace=True)

    return df_cleaned


# Apply and Save
df_enriched = add_rdkit_features(df_cleaned, cation_smiles, anion_smiles)
```
